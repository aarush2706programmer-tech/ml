```python
import pandas as pd
df = pd.read_csv('crime_incidents_messy.csv')
print(df.head())
```

```python
print(df.info())
```

```python
print(df.shape)
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()
print(df.shape)
```

```python
print(df.isnull().sum())
```

```python
df["latitude"] = df["latitude"].fillna(df["latitude"].mean())
df["longitude"] = df["longitude"].fillna(df["longitude"].mean())
print(df.isnull().sum())
```

```python
df = df.dropna(subset = ["suspect_first_name", "suspect_last_name"])
print(df)
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna(subset = ["victim_first_name", "victim_last_name", "victim_age","victim_gender"])
print(df.isnull().sum())
```

```python
df = df.dropna(subset=["suspect_age", "suspect_gender","suspect_race","victim_phone"])
print(df.isnull().sum())
```

```python
df["incident_datetime"] = df["incident_datetime"].fillna(df["incident_datetime"].mode()[0])
print(df.isnull().sum())
```

```python
df["badge_number"] = df["badge_number"].fillna(df["badge_number"].mode()[0])
print(df.isnull().sum())
```

```python
df = df.dropna(subset=["property_loss_usd", "reported_online","notes"])
print(df.isnull().sum())
```

```python
df.columns = df.columns.str.lower().str.title()
print(df.columns)
```
