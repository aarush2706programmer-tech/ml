```python
import pandas as pd
df = pd.DataFrame({
    "Name" : ["alice sharma", "bob smith", "charlie brown"],
    "Score" : [85, 90, 78],
    "City" : ["New York", "Los Angeles", "Chicago"]
})
df["Name"] = df["Name"].str.title()
print(df)
```

```python
df["Score"] = df["Score"].astype(str) + "%"
print(df)
```

```python
city_map = {"New York": "South", "Los Angeles": "West", "Chicago": "Midwest"}
df["Region"] = df["City"].map(city_map)
print(df)
```

```python
df["Pass"] = df["Score"].apply(lambda x: "Yes" if int(x.rstrip('%')) >= 80 else "No")
print(df)
```

```python
df1 = pd.DataFrame({
    "Name" : ["Aryan","Priya"],
    "Maths" : [92, 88]
})
df2 = pd.DataFrame({
    "Name" : ["Rahul","Kabir"],
    "Maths" : [75, 82],
})
combined_df = pd.concat([df1, df2], ignore_index=True)
print(combined_df)
```

```python
score = pd.DataFrame({
    "Name" : ["Aryan", "Priya", "Rahul"],
    "Maths" : [85, 90, 78],
})
details = pd.DataFrame({
    "Name" : ["Aryan", "Priya","Kabir"],
    "City": ["New York", "Los Angeles", "Chicago"]
})
inner = pd.merge(score, details, on="Name", how="inner")
print(inner)
```

```python
left = pd.merge(score, details, on="Name", how="left")
left2 = pd.merge(details, score, on="Name", how="left")
print(left)
print(" ")
print(left2)
```
