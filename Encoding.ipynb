```python
import pandas as pd
df = pd.read_csv('titanic.csv')
print(df.head())
```

```python
#Label Encoding
df["Sex"] = df["Sex"].map({'male': 0, 'female': 1})
print(df["Sex"].value_counts())
```

```python
print(df["Embarked"].value_counts())
```

```python
#pd.get_dummies function in pandas is used to convert categorical variables into dummy/indicator variables. It creates a new column for each unique value in the specified column and assigns a 1 or 0 to indicate the presence of that value in each row.
df = pd.get_dummies(df, columns=["Embarked"],drop_first=True)
print(df.columns.tolist())
```
