```python
df = pd.DataFrame({
    "Name" : ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age" : [25, 30, 35, 40, 45],
    "Maths" : [90, 80, 70, 85, 75],
    "Science" : [95, 85, 75, 80, 70],
    "English" : [88, 78, 68, 82, 72],
    "Passed" : ["Yes", "Yes", "No", "Yes", "No"]
})
#Single column - Series
print(df["Maths"])
```

```python
#Multiple columns - DataFrame
print(df[["Name", "Maths", "Passed"]])
```

```python
#All columns except one
print(df.drop("Name", axis=1))
```

```python
print(df.drop(columns=["Name"]))
```

```python
#Select only numeric columns
print(df.select_dtypes(include="number"))
```

```python
passed = df[df["Passed"] == "Yes"]
print(passed)
```

```python
high_maths = df[df["Maths"] > 80]
print(high_maths)
```

```python
top = df[(df["Maths"] > 80) & (df["Science"] > 85)]
print(top)
```

```python
selected = df[df["Name"].isin(["Alice", "David"])]
print(selected)
```

```python
not_top = df[df["Maths"]!=df["Maths"].max()]
print(not_top)
```

```python
df["Total"] = df["Maths"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3
print(df)
```

```python
df["Grade"] = df["Average"].apply(lambda x: "A" if x >= 85 else ("B" if x >= 75 else "C"))
print(df)
```

```python
df = df.rename(columns={"Passed": "Pass_Fail"})
print(df)
```

```python
print(df.groupby("Pass_Fail")["Maths"].mean())
```

```python
print(df.groupby("Grade")["Name"].count())
```

```python
print(df.groupby("Pass_Fail")["Average"].agg(["mean", "min", "max"]))
```

```python
print(df.groupby("Age")[["Maths", "Science", "English"]].mean())
```

```python
print(df["Grade"].value_counts())
```

```python
print(df.sort_values("Average", ascending=False))
```

```python
print(df.sort_values(["Age", "Average"], ascending=[True, False]))
```

```python
df_sorted = df.sort_values("Average", ascending=False).reset_index(drop=True)
print(df_sorted)
```

```python

df["Rank"] = df["Average"].rank(ascending=False).astype(int)
print(df[["Name", "Average", "Rank"]].sort_values("Rank"))
```

```python
data = {
    "Name": ["Aarav", "Vihaan", "  Anaya", "Diya", "Ishaan", "vihaan", None],
    "Age": [15, "16", 14, None, " fifteen", 16, 15],
    "Marks": [88, "90", None, 45, " 78 ", "eighty", 92],
    "City": ["Pune", "Mumbai", "pune ", None, "Delhi", "MUMBAI", "Delhi"],
    "Passed": ["Yes", "yes", "No", "no", "Yes", None, "YES"]
}

df = pd.DataFrame(data)

print(df)
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors='coerce')
print(df)
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
print(df)
```

```python
df["Marks"] = pd.to_numeric(df["Marks"], errors='coerce')
print(df)
```

```python
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
print(df)
```

```python
df["City"] = df["City"].str.strip().str.title()
print(df)
```

```python
df["City"] = df["City"].fillna(df["City"].mode()[0])
print(df)
```
