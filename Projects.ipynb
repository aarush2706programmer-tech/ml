```python
import pandas as pd
data = {
    "Name": ["Aryan", "Priya", "Kabir", "Sneha", "Rahul", "Anaya", "Rohan", "Meera"],
    
    "Maths": [88, 95, 72, 81, 67, 90, 76, 84],
    
    "Science": [92, 89, 70, 85, 74, 94, 80, 88],
    
    "English": [79, 91, 68, 87, 72, 85, 78, 90],
    
    "History": [85, 93, 75, 80, 69, 88, 77, 86]
}
df = pd.DataFrame(data)
print(df)
```

```python
df["Average"] = df[["Maths", "Science", "English", "History"]].mean(axis=1)
print(df)
```

```python
df["Total"] = df[["Maths", "Science", "English", "History"]].sum(axis=1)
print(df)
```

```python
df["Grade"] = df["Average"].apply(lambda x: 'A' if x >= 90 else ('B' if x >= 80 else ('C' if x >= 70 else 'D')))
print(df)
```

```python
df = df.sort_values(by="Average", ascending=False,ignore_index=True) 
print(df)
```

```python
df["Rank"] = df["Average"].rank(ascending=False).astype(int)
print(df)
```

```python
average_per_subject = df[["Maths", "Science", "English", "History"]].mean()
print(average_per_subject)
```

```python
#Project 2

data = {
    "Name": [
        "Virat", "Rohit", "Gill", "Rahul", "Hardik",
        "Sky", "Pant", "Jadeja", "Iyer", "Rinku"
    ],

    "Matches": [120, 135, 45, 80, 70, 60, 55, 90, 50, 35],

    "Runs": [5800, 6200, 2100, 3400, 1800, 2500, 2300, 2700, 1950, 1200],

    "HighScore": [183, 209, 128, 112, 92, 117, 146, 87, 105, 79],

    "Innings": [110, 128, 42, 75, 65, 58, 50, 82, 47, 32],

    "NotOuts": [12, 15, 4, 8, 10, 6, 5, 14, 3, 7]
}

df = pd.DataFrame(data)

print(df)
```

```python
df["Batting_Average"] = df["Runs"] / (df["Innings"] - df["NotOuts"])
print(df)
```

```python
df["Balls"] = [5000, 5500, 1800, 3000, 1500, 2000, 1900, 2200, 1600, 900]
print(df)
```

```python
df["StrkRate"] = (df["Runs"] / df["Balls"]) * 100
print(df)
```

```python
average_above_40 = df[df["Batting_Average"] > 40]
print(average_above_40["Name"])
```

```python
df["Role"] = ["Opener", "Opener", "Top-order", "Top-order", "All-rounder", "Middle-order", "Wicketkeeper", "All-rounder", "Middle-order", "Finisher"]
print(df)
```

```python
average_per_role = df.groupby("Role")["Batting_Average"].mean()
print(average_per_role)
```

```python
df["Batting_Average_Rank"] = df["Batting_Average"].rank(ascending=False).astype(int)
print(df)
```

```python
#Project D
student_scores = pd.DataFrame({
    "Name": [
        "Aryan", "Priya", "Kabir", "Sneha", "Rahul",
        "Anaya", "Rohan", "Meera", "Aditya", "Ishita"
    ],

    "Maths": [88, 95, 72, 81, 67, 90, 76, 84, 91, 78],

    "Science": [92, 89, 70, 85, 74, 94, 80, 88, 93, 82],

    "English": [79, 91, 68, 87, 72, 85, 78, 90, 88, 84]
})

print(student_scores)
```

```python
student_info = pd.DataFrame({
    "Name": [
        "Aryan", "Priya", "Kabir", "Sneha", "Rahul",
        "Anaya", "Rohan", "Meera", "Aditya", "Ishita"
    ],

    "Age": [14, 13, 14, 13, 14, 13, 14, 13, 14, 13],

    "City": [
        "Pune", "Mumbai", "Delhi", "Pune", "Bangalore",
        "Mumbai", "Delhi", "Pune", "Bangalore", "Mumbai"
    ],

    "Club": [
        "Science", "Drama", "Sports", "Science", "Music",
        "Drama", "Sports", "Science", "Music", "Drama"
    ]
})

print(student_info)
```

```python
df = pd.merge(student_scores, student_info, on="Name", how="inner")
print(df)
```

```python
df["Average"] = df[["Maths", "Science", "English"]].mean(axis=1)
print(df)
```

```python
df["Total"] = df[["Maths", "Science", "English"]].sum(axis=1)
print(df)
```

```python
df = df.sort_values(by="Average", ascending=False, ignore_index=True)
df["Club_Rank"] = df.groupby("Club")["Average"].rank(ascending=False).astype(int)
print(df)
```

```python
df["City_Rank"] = df.groupby("City")["Total"].rank(ascending=False).astype(int)
print(df)
```
