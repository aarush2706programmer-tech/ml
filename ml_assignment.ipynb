```python
marks = np.array([
    [88, 92, 79, 85],   # Student 0
    [95, 87, 91, 78],   # Student 1
    [72, 68, 75, 70],   # Student 2
    [45, 52, 60, 55],   # Student 3
    [91, 88, 84, 93],   # Student 4
    [63, 71, 58, 66],   # Student 5
])
print(marks.shape)
```

```python
print(marks.mean(axis=0))
```

```python
print(marks.mean(axis=1))
```

```python
print(marks.mean(axis=1).argmax())
```

```python
print(marks[4,:])
```

```python
print(np.where((marks < 60).any(axis=1))[0])
```

```python
scaled_array = marks / 10
print(scaled_array)
```

```python
print(np.round(scaled_array, 1))
```
