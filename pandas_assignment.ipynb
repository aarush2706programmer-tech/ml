```python
import pandas as pd
products = pd.DataFrame({
    'Product': ['Pen','Book','Ruler','Bag','Pencil','Eraser'],
    'Price':   [10, 150, 25, 500, 5, 8],
    'Stock':   [200, 50, 80, 30, 300, 150],
    'Category':['Stationery','Stationery','Stationery','Bag','Stationery','Stationery'],
})
print(products[products["Price"]>50])
```

```python
products["Value"] = products["Price"] * products["Stock"]
print(products)
```

```python
print(products.groupby("Category")["Value"].sum())
```

```python
products = products.sort_values("Value", ascending=False)
print(products)
```

```python
print(products.head(3))
```

```python
products["Expensive"] = products["Price"].apply(lambda x: True if x > 100 else False)
print(products)
```

```python
messy = pd.DataFrame({
    'Student': ['Aryan', 'Priya', None, 'Kabir', 'Aryan'],
    'Score':   [88, 95, 72, None, 88],
    'Grade':   ['B', 'A', 'C', 'B', 'B'],
    'City':    ['Pune', 'delhi', 'PUNE', 'Mumbai', 'Pune'],
})
print(messy)
```

```python
messy = messy.drop_duplicates()
print(messy)
```

```python
messy["Score"] = messy["Score"].fillna(messy["Score"].mean())
print(messy)
```

```python
messy["City"] = messy["City"].str.strip().str.title()
print(messy)
```

```python
messy["Student"] = messy["Student"].fillna("Unknown")
print(messy)
```
